#Código para carregar Data.txt do X-Plane

**Importar bibliotecas**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Carregar arquivo**

```python
file_path = "Data.txt"

with open(file_path, "r", encoding="utf-8") as file:
    cleaned_content = [line.replace("\x00", "").strip() for line in file if line.strip()]

# Criar DataFrame separando os campos pelo delimitador '|'
df = pd.DataFrame([line.split("|") for line in cleaned_content])

# Definir os valores da primeira linha como os novos nomes das colunas
df.columns = df.iloc[0]

# Remover a primeira linha, pois agora ela está duplicada como cabeçalho
df = df.drop(0)

# Remover espaços em branco dos nomes das colunas
df.columns = df.columns.str.strip()

# Deletar a última coluna sem valores
df = df.drop(df.columns[-1], axis=1)

# Converter para valor numérico
df = df.apply(pd.to_numeric, errors='coerce')

# Exibir as primeiras linhas do DataFrame
df
```

**Lista de colunas (variáveis)**

```python
print(df.columns)
```

**Para visualizar a plotagem da coluna, altere col_index**

```python
col_index = "power,_1,hp"

# Gerar o gráfico temporal
df[col_index].plot(figsize=(12, 6))

# Adicionar rótulos aos eixos
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title(f"Gráfico Temporal da Coluna: {col_index}")

# Exibir o gráfico
plt.show()
```
